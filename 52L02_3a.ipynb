```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
x = np.arange(-1,1,0.01)
h11 = 1/(1+np.exp(-((500*x)+30)))
h12 = 1/(1+np.exp(-((500*x)-30)))
h21=h11-h12
```

```python
plt.title("GRAPH FOR THE FUNCTION h11(x)")
plt.plot(x,h11)
plt.ylabel("$h_{11}(x)$")
plt.show()
```

```python
plt.title("GRAPH FOR THE FUNCTION h12(x")
plt.plot(x,h12)
plt.ylabel("$h_{11}(x)$")
plt.show()
```

```python
plt.title("GRAPH FOR THE FUNCTION h21(x)")
plt.ylabel("$h_{21}(x)$")
plt.plot(x,h21)
plt.show()
```
